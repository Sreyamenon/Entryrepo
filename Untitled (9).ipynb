Real-world healthcare data may contain missing values. Missing data can affect analysis accuracy, so I first check how many missing values are present in each column.

```python
df.isnull().sum()
```

Duplicate records can lead to biased results by over-representing certain patients.
I first check for duplicates and then remove them.

Text data may contain inconsistent capitalization or extra spaces.
I standardize important text columns to ensure consistency during grouping and analysis.

```python
text_cols = ['Name', 'Doctor', 'Hospital', 'Medical Condition']

for col in text_cols:
    df[col] = df[col].str.title().str.strip()
```

Date columns must be converted to datetime format to perform time-based analysis and calculations such as hospital stay duration. and one more column added to calculate the length of the stay

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

I re-check data types to confirm numerical and date columns are correctly formatted before performing analysis and visualization.

```python
df.dtypes
```

In real-world healthcare data, duplicate records may exist due to repeated entries or system errors.
Duplicate rows can bias analysis by over-representing certain patients or treatments.

Therefore, I first identify and then remove duplicate records.

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

For better readability and presentation, I rearrange the columns so that:

Patient details come first

Medical and hospital information comes next

Billing and derived columns come at the end


```python
new_order = [
    'Name', 'Age', 'Gender', 'Blood Type',
    'Medical Condition', 'Admission Type',
    'Date of Admission', 'Discharge Date', 'Length of Stay',
    'Doctor', 'Hospital', 'Medication',
    'Insurance Provider', 'Billing Amount', 'Room Number'
]

df = df[[col for col in new_order if col in df.columns]]
```

Certain values are logically impossible in a healthcare dataset:

Age = 0 or negative

Billing Amount = 0

Length of stay = 0 (in many hospital cases)

I check for such values to ensure data validity.

```python
df[(df['Age'] <= 0) | (df['Billing Amount'] <= 0)]
```

Records with impossible values can distort analysis and lead to incorrect conclusions.
Therefore, I remove such records from the dataset.

```python
df = df[(df['Age'] > 0) & (df['Billing Amount'] > 0)]
```

Filtering allows focused analysis.
For example, I may want to analyze only Emergency admissions or patients above a certain age.

```python
emergency_df = df[df['Admission Type'] == 'Emergency']
```

```python
senior_patients = df[df['Age'] >= 60]
```

created one more column for age group

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 18, 35, 60, 120],
    labels=['Child', 'Young Adult', 'Adult', 'Senior']
)
```

```python
df.columns
```

```python
df.groupby('Age_Group')['Length of Stay'].mean()
```

Creating new features improves analysis depth.
Derived columns convert raw data into meaningful healthcare indicators.

```python
df['Billing_Category'] = pd.cut(
    df['Billing Amount'],
    bins=[0, 20000, 50000, df['Billing Amount'].max()],
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Admission_Month'] = df['Date of Admission'].dt.month
```

```python
df['Stay_Category'] = pd.cut(
    df['Length of Stay'],
    bins=[0, 3, 7, 14, df['Length of Stay'].max()],
    labels=['Short', 'Medium', 'Long', 'Very Long']
)
```

DATA VISUALIZATION

Age Distribution (Histogram) - This histogram shows the distribution of patients across different age groups. It helps understand which age range dominates hospital admissions.


```python
plt.figure()
plt.hist(df['Age'], bins=10)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

Gender Distribution (Bar Chart) - This bar chart compares the number of male and female patients admitted to the hospital.

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

Blood Type Distribution (Pie Chart) - This pie chart shows the proportion of patients by blood group, which is useful for medical inventory planning.

```python
df['Blood Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Blood Type Distribution')
plt.ylabel('')
plt.show()
```

Medical Condition Frequency (Bar Chart) This chart displays how frequently each medical condition appears in the dataset.

```python
df['Medical Condition'].value_counts().plot(kind='bar')
plt.title('Medical Condition Frequency')
plt.xlabel('Medical Condition')
plt.ylabel('Number of Patients')
plt.xticks(rotation=45)
plt.show()
```

Admission Type Distribution (Count Plot) - This plot shows how patients are distributed across different admission types such as Emergency and Elective.

```python
sns.countplot(x='Admission Type', data=df)
plt.title('Admission Type Distribution')
plt.show()
```

Billing Amount Distribution (Histogram) - This histogram shows how treatment costs are distributed across patients.

```python
plt.figure()
plt.hist(df['Billing Amount'], bins=10)
plt.title('Billing Amount Distribution')
plt.xlabel('Billing Amount')
plt.ylabel('Frequency')
plt.show()
```

Billing Amount by Medical Condition (Box Plot) - This box plot compares treatment costs across different medical conditions and highlights cost variations and outliers.

```python
sns.boxplot(x='Medical Condition', y='Billing Amount', data=df)
plt.xticks(rotation=45)
plt.title('Billing Amount by Medical Condition')
plt.show()
```

Length of Stay by Admission Type (Box Plot) - This plot compares hospital stay duration across different admission types.

```python
sns.boxplot(x='Admission Type', y='Length of Stay', data=df)
plt.title('Length of Stay by Admission Type')
plt.show()
```

Insurance Provider Distribution (Bar Chart) - This visualization shows how patients are distributed across different insurance providers.

```python
df['Insurance Provider'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Insurance Provider Distribution')
plt.show()

```

Admissions Over Time (Line Chart) - This line chart shows the trend of patient admissions over time based on admission dates.
